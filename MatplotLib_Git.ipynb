# Quick Exercise!

```python
data = pd.read_csv('gapminder.csv')
df = data.copy()
```

Build scatter plot of GDP per Capita vs Life Expentancy for year 1970 and specify Pakistan, UAE and UK.

```python
df.head()
```

```python
dict = {
    'Pakistan':'green',
    'United Kingdom':'yellow',
    'United Arab Emirates':'red',
}

col = {'Pakistan':'green','United Arab Emirates':'yellow', 'United Kingdom':'red'}
```

```python
data = df[(df['Country'] == 'Pakistan') | (df['Country'] == 'United Kingdom') | (df['Country'] == 'United Arab Emirates')]
```

```python
data.head()
```

```python
data['color'] = data['Country'].map(col)
```

```python
data.tail(2)
```

```python
plt.figure(figsize = (15, 5))

plt.scatter(data['gdp'], data['life'])
plt.xscale('log')
plt.xlabel('GDP per capita [in USD]')
plt.ylabel('Life expectancy [in years]')
plt.title('World Development')

plt.show()
```

```python
data['cal_pop'] = data['population'] / 1000000
pop = data['cal_pop']
pop = pop * 4
```

```python
plt.figure(figsize = (15, 5))

plt.scatter(data['gdp'], data['life'], s = pop, color = data['color'], alpha = 0.5)
plt.xscale('log')
plt.xlabel('GDP per capita [in USD]')
plt.ylabel('Life expectancy [in years]')
plt.title('World Development')

plt.show()
```

```python
plt.figure(figsize = (15, 5))

plt.scatter(data['gdp'], data['life'], s = pop, color = data['color'], alpha = 0.5)
plt.xscale('log')
plt.xlabel('GDP per capita [in USD]')
plt.ylabel('Life expectancy [in years]')
plt.title('World Development')

plt.text(3900,69,'Pakistan')

plt.grid(True)
plt.show()
```
